5. Scenario: A study is conducted to compare the performance of two different teaching methods. The test scores of the students in each group are as follows:
   Group A: [85, 90, 92, 88, 91]
   Group B: [82, 88, 90, 86, 87]
   Perform a t-test to determine if there is a significant difference in the mean scores between the two groups.

To perform a t-test and determine if there is a significant difference in the mean scores between Group A and Group B, we can use a two-sample independent t-test. This test compares the means of two independent groups and assesses whether the difference between the means is statistically significant.

The test scores for each group are as follows:
Group A: [85, 90, 92, 88, 91]
Group B: [82, 88, 90, 86, 87]

Here's how you can perform a t-test in Python using the scipy library:
import scipy.stats as stats

group_a = [85, 90, 92, 88, 91]
group_b = [82, 88, 90, 86, 87]

# Perform independent t-test
t_stat, p_value = stats.ttest_ind(group_a, group_b)

# Print the results
print("T-statistic:", t_stat)
print("p-value:", p_value)
The output will give you the t-statistic and the p-value. The t-statistic measures the difference between the means of the two groups, while the p-value indicates the probability of observing such a difference by chance.

In this case, if the p-value is below a certain significance level (commonly 0.05), we can reject the null hypothesis, which suggests that there is no significant difference between the means. If the p-value is above the significance level, we fail to reject the null hypothesis, indicating that there is not enough evidence to conclude a significant difference.

13. Scenario: A survey is conducted to measure customer satisfaction with a product on a scale of 1 to 10. The data collected is as follows:
    [7, 9, 6, 8, 10, 7, 8, 9, 7, 8]
    Calculate the 95% confidence interval for the population mean satisfaction score.

To calculate the 95% confidence interval for the population mean satisfaction score, we can use the following formula:

Confidence Interval = [sample mean - (critical value * standard error), sample mean + (critical value * standard error)]

The data collected for customer satisfaction scores on a scale of 1 to 10 is as follows:
[7, 9, 6, 8, 10, 7, 8, 9, 7, 8]
import numpy as np
import scipy.stats as stats

satisfaction_scores = [7, 9, 6, 8, 10, 7, 8, 9, 7, 8]

# Calculate sample mean and standard deviation
sample_mean = np.mean(satisfaction_scores)
sample_std = np.std(satisfaction_scores, ddof=1)  # ddof=1 for sample standard deviation

# Calculate standard error
standard_error = sample_std / np.sqrt(len(satisfaction_scores))

# Calculate the critical value (z-score) for 95% confidence level
critical_value = stats.t.ppf(0.975, df=len(satisfaction_scores)-1)

# Calculate the confidence interval
confidence_interval = [sample_mean - (critical_value * standard_error),
                      sample_mean + (critical_value * standard_error)]

# Print the confidence interval
print("95% Confidence Interval:", confidence_interval)
The output will provide the 95% confidence interval for the population mean satisfaction score.

Please note that we used the t-distribution to calculate the critical value since the sample size is small (less than 30). If the sample size is large (typically 30 or more), you can use the standard normal distribution (Z-distribution) instead by replacing stats.t.ppf(0.975, df=len(satisfaction_scores)-1) with stats.norm.ppf(0.975) in the code.

import numpy as np

satisfaction_scores = [7, 8, 9, 6, 8, 7, 9, 7, 8, 7]

# Calculate the standard deviation
std_dev = np.std(satisfaction_scores, ddof=1)  # ddof=1 for sample standard deviation

# Calculate the sample size
sample_size = len(satisfaction_scores)

# Calculate the standard error of the mean
standard_error = std_dev / np.sqrt(sample_size)

# Print the standard error of the mean
print("Standard Error of the Mean:", standard_error)

The standard error of the mean quantifies the variability of sample means around the true population mean. It represents the average amount by which the sample means would vary across repeated sampling.

20. Scenario: A company wants to analyze the relationship between advertising expenditure and sales. The data collected is as follows:
    Advertising Expenditure (in thousands): [10, 15, 12, 8, 14]
    Sales (in thousands): [25, 30, 28, 20, 26]
    Perform a multiple regression analysis to predict sales based on advertising expenditure.

To perform a multiple regression analysis to predict sales based on advertising expenditure, we can fit a linear regression model to the data using the least squares method. Multiple regression allows us to analyze the relationship between multiple independent variables and a dependent variable.

The data collected for advertising expenditure (in thousands) and sales (in thousands) is as follows:
Advertising Expenditure: [10, 15, 12, 8, 14]
Sales: [25, 30, 28, 20, 26]

import numpy as np
import statsmodels.api as sm

advertising_expenditure = [10, 15, 12, 8, 14]
sales = [25, 30, 28, 20, 26]

# Add a constant term to the advertising expenditure
X = sm.add_constant(advertising_expenditure)

# Fit the multiple regression model
model = sm.OLS(sales, X)
results = model.fit()

# Print the results
print(results.summary())
The output will provide the summary of the multiple regression analysis, including the coefficients, standard errors, t-values, and p-values for the independent variables.
